```python
import numpy as np
import pandas as pd
import math
import sys
```

```python
# subir angulos y distancia paara regresion
df = pd.read_excel('angulos y distancias para regresion 10 casos brazo azul.xlsx')
df1 = pd.read_excel('angulo S0 para regresion brazo azul.xlsx')
print(df.head())
print(df.shape)
print(df1.head())
print(df1.shape)
```

```python
# Create separate object for target variable

y1 = df.S1
y1 = y1.as_matrix()

y2 = df.S2
y2 = y2.as_matrix()

y3 = df.S3
y3 = y3.as_matrix()

y0 = df1.S0
y0 = y0.as_matrix()

# Create separate object for input features
X = df.distancia
X = X.as_matrix()
X = X.reshape(-1, 1)

# crear X paraS0
X0 = df1.angulo
X0 = X0.as_matrix()
X0 = X0.reshape(-1, 1)
```

```python
from sklearn.tree import DecisionTreeRegressor

```

```python
regr_1 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(X, y1)
predictions1 = regr_1.predict(X)

regr_2 = DecisionTreeRegressor(max_depth=5)
regr_2.fit(X, y2)
predictions2 = regr_2.predict(X)

regr_3 = DecisionTreeRegressor(max_depth=5)
regr_3.fit(X, y3)
predictions3 = regr_3.predict(X)

regr_0 = DecisionTreeRegressor(max_depth=6)
regr_0.fit(X0, y0)
predictions0 = regr_0.predict(X0)
```

```python
import matplotlib.pyplot as plt
plt.plot(y1)
plt.plot(predictions1)
plt.show()
```

```python
plt.plot(y2)
plt.plot(predictions2)
plt.show()
```

```python
plt.plot(y3)
plt.plot(predictions3)
plt.show()
```

```python
plt.plot(y0)
plt.plot(predictions0)
plt.show()
```

```python
# Entrar las coordenadas para estimar los angulos de los 4 DOF
coordx = 35
coordy = 14

# Predecir angulos de los 4 DOF a partir de coordenadas X y Y
para_dist = ((coordx-20)*(coordx-20)) + coordy*coordy
distancia = np.array(math.sqrt(para_dist)).reshape(1, -1)

cos_angulo = (coordx-20)/distancia
angulo = math.degrees(math.acos(cos_angulo))

S1 = regr_1.predict(distancia)
S2 = regr_2.predict(distancia)
S3 = regr_3.predict(distancia)
S0 = regr_0.predict(angulo)

print('Angulo teorico = ',angulo)
print('Distancia = ',distancia)
print('S0 = ',S0.round(0))
print('S1 = ',S1.round(0))
print('S2 = ',S2.round(0))
print('S3 = ',S3.round(0))
```
